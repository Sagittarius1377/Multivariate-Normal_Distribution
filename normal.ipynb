# Normal Distribution 

Distribution where most data points cluster around central mean value , tapering off equally on both side 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 65
std  = 5
```

```python
weight = np.random.normal(loc=mean,scale=std,size=280000000)
```

```python

bins = 500
```

```python
min_weight = np.min(weight)
max_weight = np.max(weight)
bin_width = (max_weight-min_weight)/bins
```

```python
print(weight)
```

```python
print(min_weight)
print(max_weight)
```

```python
print(bin_width)
```

```python
plt.xlabel("Weights in KG")
plt.ylabel("Probability")
plt.hist(weight,bins=bins)
plt.grid(False)
```
